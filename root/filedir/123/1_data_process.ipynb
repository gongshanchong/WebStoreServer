 XX[:] = X + Y  X += Y 

```python
import torch

x = torch.arange(12,dtype=torch.float32).reshape((3,4))
y = torch.tensor([[2.0,1,4,3],[1,2,3,4],[4,3,2,1]])
z = torch.zeros_like(y) # z y0
print("id(z):",id(z))    
z[:] = x + y
print("id(z):",id(z))   
```

```python
import torch

x = torch.arange(12,dtype=torch.float32).reshape((3,4))
y = torch.tensor([[2.0,1,4,3],[1,2,3,4],[4,3,2,1]])
before = id(x) 
x += y
print(id(x) == before)
```

```python
#inputs  outputs 
# pandas
# !pip install pandas
import pandas as pd
import os

os.makedirs(os.path.join('.','01_Data'),exist_ok=True) # 
data_file = os.path.join('.','01_Data','01_house_tiny.csv')
with open(data_file,'w') as f:
    f.write('NumRooms,Alley,Price\n') # 
    f.write('NA,Pave,127500\n')  # 
    f.write('2,NA,106000\n')
    f.write('4,NA,178100\n')
    f.write('NA,NA,140000\n')
    
data_file = os.path.join('.','01_Data','01_house_tiny.csv')
data = pd.read_csv(data_file)

inputs, outputs = data.iloc[:,0:2], data.iloc[:,2]
inputs = inputs.fillna(inputs.mean()) #  NaN 
inputs = pd.get_dummies(inputs, dummy_na=True) # dummy_na=True  Alley10   
print(inputs)

x, y = torch.tensor(inputs.values), torch.tensor(outputs.values)
print(x)
print(y)
```
